```python
y = load_airline()
y.head()
```

## Moving average for extracting trend

* Data has a monthly granularity and yearly seasonality, so lets use a window = 12
* When window size is even
    * Do not center the window
    * Perform an additional 2-MA 
    * Shift backwards to aling the index `shift(-window-size//2)`

```python
trend = (y.rolling(window=12).mean().rolling(window=2).mean().shift(periods=-6))
```

## Visualize

```python
plot_series(y, trend, labels=['series', 'trend']);
```
